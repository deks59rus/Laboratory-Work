1.3 Исследуйте, в каких столбцах таблиц содержатся пропуски. Посчитайте долю строк, содержащих пропуски, в отношении к общему количеству строк.

```python
def NanCounter(table):
    t1=pd.DataFrame(table)
    t1= t1.isna().sum(axis=1)
    t1_num_rows = len(t1[t1 != 0].index)
    print("Кол-во пропусков в таблице по столбцам:\n%s" % (table.isna().sum()))
    print("%3.5f%% доля строк в таблице содержащих пропуски" % (t1_num_rows/t1.shape[0]*100))
    
print("Для DataFrame 'Recipes':" )
NanCounter(recipes)
print()
print("Для DataFrame 'Reviews':" )
NanCounter(reviews)
```
